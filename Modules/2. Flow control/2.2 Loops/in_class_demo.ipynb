# For loops
For loops in python are used to iterate over:
-  a sequence (lists, tuples, string)
- a range of values (ex. 1 to 100)
- any other iterable object in python (dictionaries, sets, generators). 

Iterating over a sequence is called a traversal, so you can think of a for loop as stepping through or *traversing* a sequence of items.

# The `range()` method
Python has a function called `range()` that is used to generate a range of numbers. 

The range function can take up to 3 arguments: `range(start, stop, step_size)`
- `start`: lets the function know what value to start at
- `stop`: lets the function know which value to end on (this value is not included)
- `step`: lets the function know how many values to count at once (default is 1)

You can always know how many values will be generated by subtracting the starting value from the stop value.

`range(25, 100) -> 100 - 25 = 75 values total`

**Lets see how the range function works before we use it in a loop**

```python
#Range by itself wont give us a list but instead a range object
print(type(range(10)))
# print(range(10))

# in order to see the actual values we need to cast it to the list type
# print(list(range(10)))
```

# Looping over a range of values 
To loop over a range of values is almost identical to looping through a sequence of values, except we replace the `sequence` with the `range()` method. 
<p><b style = "color: purple">for</b> value <b style = "color: purple">in</b> range(10):</p>

      `loop body`

but now instead of looping over each value in a sequence, we are looping over the values returned by the range function.

```python
for value in range(10):
    print(value, end = " ")
```

**The advantage of using this method is that now we can use the values to index a list and also know what iteration we are on**

```python
var = ['Hello', "I am a list", True, False, (1,2,3,4)]
for value in range(len(var)):
    print(f"{value}) {var[value]}")
```
